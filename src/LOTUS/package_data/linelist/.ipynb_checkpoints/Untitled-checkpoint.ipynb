```python
import pandas as pd
import numpy as np
```

```python
linelist = pd.read_csv("final.linelist")
```

```python
linelist
```

```python
newlinelist = linelist[~linelist.duplicated(subset=['Line','lambda', "Ex.P"])]
```

```python
idx_fei = np.where(newlinelist.Line=="FeI")
newlinelist_fei = newlinelist.iloc[idx_fei].sort_values(by=["lambda"])
```

```python
idx_feii = np.where(newlinelist.Line=="FeII")
newlinelist_feii = newlinelist.iloc[idx_feii].sort_values(by=["lambda"])
```

```python
newlinelist_feii
```

```python
finalnewlinelist = newlinelist_fei.append(newlinelist_feii, ignore_index=True)
finalnewlinelist = finalnewlinelist[["Line", "lambda", "Ex.P"]]
```

```python
finalnewlinelist
```

```python
finalnewlinelist.to_csv("newfinal.linelist", index=False)
```
